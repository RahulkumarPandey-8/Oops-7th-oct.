# **Problem 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.**


```python
class BankAccount:
    def __init__(self, account_number, account_holder, balance=0):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ${amount}. New balance: ${self.balance}")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.balance}")
        else:
            print("Insufficient funds.")

    def display_balance(self):
        print(f"Account Balance for {self.account_holder}: ${self.balance}")

# Function to interactively get user input
def get_user_input():
    account_number = input("Enter account number: ")
    account_holder = input("Enter account holder name: ")
    initial_balance = float(input("Enter initial balance (default is 0): ") or 0)

    return BankAccount(account_number, account_holder, initial_balance)

# Example usage with user input
account = get_user_input()
account.display_balance()

deposit_amount = float(input("Enter deposit amount: "))
account.deposit(deposit_amount)

withdraw_amount = float(input("Enter withdrawal amount: "))
account.withdraw(withdraw_amount)
```

# **Problem 2: Employee Management Create a class representing an employee with attributes like employee ID, name, and salary. Implement methods to calculate the yearly bonus and display employee details.**

```python
class Employee:
    def __init__(self, employee_id, name, salary):
        self.employee_id = employee_id
        self.name = name
        self.salary = salary

    def calculate_yearly_bonus(self):
        # Assuming a fixed bonus rate of 10%
        return 0.1 * self.salary

    def display_details(self):
        print(f"Employee ID: {self.employee_id}")
        print(f"Name: {self.name}")
        print(f"Salary: ${self.salary}")


class EmployeeManagement:
    def __init__(self):
        self.employees = []

    def add_employee(self, employee):
        self.employees.append(employee)
        print(f"Employee {employee.name} added to the system.")

    def add_employee_from_input(self):
        employee_id = input("Enter Employee ID: ")
        name = input("Enter employee name: ")
        salary = float(input("Enter employee salary: "))

        new_employee = Employee(employee_id, name, salary)
        self.add_employee(new_employee)

    def display_all_employees(self):
        print("Employee Details:")
        for employee in self.employees:
            employee.display_details()
            print()  # Add a newline for better readability

    def calculate_all_bonuses(self):
        print("Yearly Bonuses:")
        for employee in self.employees:
            bonus = employee.calculate_yearly_bonus()
            print(f"{employee.name}: ${bonus}")
            print()  # Add a newline for better readability


# Example usage:
employee_management = EmployeeManagement()

# Adding employees using user input
employee_management.add_employee_from_input()
employee_management.add_employee_from_input()

# Displaying employee details
employee_management.display_all_employees()

# Calculating and displaying yearly bonuses
employee_management.calculate_all_bonuses()
```

# **Problem 3: Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle, return a vehicle, and display available vehicles.**

```python
class Vehicle:
    def __init__(self, vehicle_id, vehicle_type):
        self.vehicle_id = vehicle_id
        self.vehicle_type = vehicle_type
        self.is_rented = False

    def rent_vehicle(self):
        if not self.is_rented:
            self.is_rented = True
            print(f"Rented {self.vehicle_type} with ID {self.vehicle_id}.")
        else:
            print(f"{self.vehicle_type} with ID {self.vehicle_id} is already rented.")

    def return_vehicle(self):
        if self.is_rented:
            self.is_rented = False
            print(f"Returned {self.vehicle_type} with ID {self.vehicle_id}.")
        else:
            print(f"{self.vehicle_type} with ID {self.vehicle_id} is not currently rented.")


class RentalSystem:
    def __init__(self):
        self.vehicles = []

    def add_vehicle(self, vehicle):
        self.vehicles.append(vehicle)
        print(f"{vehicle.vehicle_type} with ID {vehicle.vehicle_id} added to the system.")

    def rent_vehicle_from_input(self):
        vehicle_id = input("Enter Vehicle ID to rent: ")
        vehicle = next((v for v in self.vehicles if v.vehicle_id == vehicle_id), None)
        if vehicle:
            vehicle.rent_vehicle()
        else:
            print(f"Vehicle with ID {vehicle_id} not found.")

    def return_vehicle_from_input(self):
        vehicle_id = input("Enter Vehicle ID to return: ")
        vehicle = next((v for v in self.vehicles if v.vehicle_id == vehicle_id), None)
        if vehicle:
            vehicle.return_vehicle()
        else:
            print(f"Vehicle with ID {vehicle_id} not found.")

    def display_available_vehicles(self):
        print("Available Vehicles:")
        for vehicle in self.vehicles:
            if not vehicle.is_rented:
                print(f"{vehicle.vehicle_type} - ID: {vehicle.vehicle_id}")


# Example usage:
vehicle1 = Vehicle("V001", "Car")
vehicle2 = Vehicle("V002", "Bike")
vehicle3 = Vehicle("V003", "Scooter")

rental_system = RentalSystem()
rental_system.add_vehicle(vehicle1)
rental_system.add_vehicle(vehicle2)
rental_system.add_vehicle(vehicle3)

rental_system.display_available_vehicles()

rental_system.rent_vehicle_from_input()
rental_system.return_vehicle_from_input()

rental_system.display_available_vehicles()
```

# **Problem 4: Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.**

```python
class Book:
    def __init__(self, book_id, title, author):
        self.book_id = book_id
        self.title = title
        self.author = author
        self.is_borrowed = False

    def borrow_book(self):
        if not self.is_borrowed:
            self.is_borrowed = True
            print(f"Borrowed the book '{self.title}' by {self.author}.")
        else:
            print(f"The book '{self.title}' by {self.author} is currently borrowed.")

    def return_book(self):
        if self.is_borrowed:
            self.is_borrowed = False
            print(f"Returned the book '{self.title}' by {self.author}.")
        else:
            print(f"The book '{self.title}' by {self.author} is not currently borrowed.")


class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)
        print(f"Book '{book.title}' by {book.author} added to the library.")

    def borrow_book_from_input(self):
        book_id = input("Enter Book ID to borrow: ")
        book = next((b for b in self.books if b.book_id == book_id), None)
        if book:
            book.borrow_book()
        else:
            print(f"Book with ID {book_id} not found.")

    def return_book_from_input(self):
        book_id = input("Enter Book ID to return: ")
        book = next((b for b in self.books if b.book_id == book_id), None)
        if book:
            book.return_book()
        else:
            print(f"Book with ID {book_id} not found.")

    def display_available_books(self):
        print("Available Books:")
        for book in self.books:
            if not book.is_borrowed:
                print(f"{book.title} by {book.author} - ID: {book.book_id}")


# Example usage:
book1 = Book("B001", "Python Basics", "John Doe")
book2 = Book("B002", "Data Structures", "Jane Smith")
book3 = Book("B003", "Machine Learning", "Alice Johnson")

library = Library()
library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

library.display_available_books()

library.borrow_book_from_input()
library.return_book_from_input()

library.display_available_books()
```

# **Problem 5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.**

```python
# Create a Product class
class Product:
    def __init__(self, product_id, name, price, quantity):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.quantity = quantity

# Create an Inventory class
class Inventory:
    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)
        print(f"{product.name} added to inventory.")

    def update_quantity(self, product_id, new_quantity):
        for product in self.products:
            if product.product_id == product_id:
                product.quantity = new_quantity
                print(f"Quantity of {product.name} updated to {new_quantity}.")
                return
        print(f"Product with ID {product_id} not found in inventory.")

    def display_products(self):
        print("Available Products:")
        print("-------------------")
        for product in self.products:
            print(f"ID: {product.product_id}, Name: {product.name}, Price: ${product.price}, Quantity: {product.quantity}")

# Function to take input from the user and perform actions
def main():
    inventory = Inventory()

    while True:
        print("\nMenu:")
        print("1. Add Product to Inventory")
        print("2. Update Product Quantity")
        print("3. Display Available Products")
        print("4. Exit")

        choice = input("Enter your choice (1/2/3/4): ")

        if choice == '1':
            product_id = input("Enter Product ID: ")
            name = input("Enter Product Name: ")
            price = float(input("Enter Product Price: "))
            quantity = int(input("Enter Product Quantity: "))
            new_product = Product(product_id, name, price, quantity)
            inventory.add_product(new_product)

        elif choice == '2':
            product_id = input("Enter Product ID to update quantity: ")
            new_quantity = int(input("Enter new quantity: "))
            inventory.update_quantity(product_id, new_quantity)

        elif choice == '3':
            inventory.display_products()

        elif choice == '4':
            print("Exiting program.")
            break

        else:
            print("Invalid choice. Please enter a valid option.")

# Run the main function
if __name__ == "__main__":
    main()
```

# **Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.**

```python
# Define Shape class
class Shape:
    def __init__(self, length=0, width=0, height=0):
        self.length = length
        self.width = width
        self.height = height

    def calculate_area(self):
        raise NotImplementedError("Subclasses should implement this method")

    def calculate_perimeter(self):
        raise NotImplementedError("Subclasses should implement this method")

# Define Rectangle class as a subclass of Shape
class Rectangle(Shape):
    def calculate_area(self):
        return self.length * self.width

    def calculate_perimeter(self):
        return 2 * (self.length + self.width)

# Define Triangle class as a subclass of Shape
class Triangle(Shape):
    def calculate_area(self):
        # Using Heron's formula for the area of a triangle
        s = (self.length + self.width + self.height) / 2
        return (s * (s - self.length) * (s - self.width) * (s - self.height))**0.5

    def calculate_perimeter(self):
        return self.length + self.width + self.height

# Function to take input from the user and create a shape
def create_shape():
    print("Choose a shape:")
    print("1. Rectangle")
    print("2. Triangle")

    choice = input("Enter your choice (1/2): ")

    if choice == '1':
        length = float(input("Enter the length of the rectangle: "))
        width = float(input("Enter the width of the rectangle: "))
        return Rectangle(length, width)

    elif choice == '2':
        length = float(input("Enter the length of the triangle: "))
        width = float(input("Enter the width of the triangle: "))
        height = float(input("Enter the height of the triangle: "))
        return Triangle(length, width, height)

    else:
        print("Invalid choice. Defaulting to Rectangle.")
        return Rectangle()

# Main function to run the program
def main():
    shape = create_shape()

    print("\nCalculations:")
    print(f"Area: {shape.calculate_area()}")
    print(f"Perimeter: {shape.calculate_perimeter()}")

# Run the main function
if __name__ == "__main__":
    main()
```

# **Problem 7: Student Management Create a class representing a student with attributes like student ID, name, and grades. Implement methods to calculate the average grade and display student details.**

```python
# Define Student class
class Student:
    def __init__(self, student_id, name):
        self.student_id = student_id
        self.name = name
        self.grades = []

    def add_grade(self, grade):
        self.grades.append(grade)

    def calculate_average_grade(self):
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)

    def display_details(self):
        print("Student Details:")
        print(f"ID: {self.student_id}")
        print(f"Name: {self.name}")
        print(f"Grades: {', '.join(map(str, self.grades))}")
        print(f"Average Grade: {self.calculate_average_grade()}")

# Function to take input from the user and create a student
def create_student():
    student_id = input("Enter Student ID: ")
    name = input("Enter Student Name: ")

    student = Student(student_id, name)

    num_grades = int(input("Enter the number of grades to input: "))
    for _ in range(num_grades):
        grade = float(input("Enter a grade: "))
        student.add_grade(grade)

    return student

# Main function to run the program
def main():
    student = create_student()
    student.display_details()

# Run the main function
if __name__ == "__main__":
    main()
```

# **Problem 8: Email Management Create a class representing an email with attributes like sender, recipient, and subject. Implement methods to send an email and display email details.**

```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

# Define Email class
class Email:
    def __init__(self, sender, recipient, subject):
        self.sender = sender
        self.recipient = recipient
        self.subject = subject
        self.content = None

    def send_email(self, content):
        self.content = content

        # Set up the MIMEText object
        message = MIMEMultipart()
        message['From'] = self.sender
        message['To'] = self.recipient
        message['Subject'] = self.subject
        message.attach(MIMEText(self.content, 'plain'))

        # Replace these values with your own SMTP server details
        smtp_server = 'your_smtp_server'
        smtp_port = 587
        smtp_username = 'your_username'
        smtp_password = 'your_password'

        # Connect to the SMTP server and send the email
        try:
            with smtplib.SMTP(smtp_server, smtp_port) as server:
                server.starttls()
                server.login(smtp_username, smtp_password)
                server.sendmail(self.sender, self.recipient, message.as_string())
            print("Email Sent!")
        except Exception as e:
            print(f"Error sending email: {e}")

    def display_details(self):
        print("Email Details:")
        print(f"Sender: {self.sender}")
        print(f"Recipient: {self.recipient}")
        print(f"Subject: {self.subject}")
        print("Content:")
        print(self.content)

# Function to take input from the user and create an email
def create_email():
    sender = input("Enter Sender's Email: ")
    recipient = input("Enter Recipient's Email: ")
    subject = input("Enter Email Subject: ")

    email = Email(sender, recipient, subject)

    content = input("Enter Email Content: ")
    email.send_email(content)

    return email

# Main function to run the program
def main():
    email = create_email()
    email.display_details()

# Run the main function
if __name__ == "__main__":
    main()
```


# **Problem 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts. Implement methods to add posts, display posts, and search for posts by keyword.**

```python
3# Define SocialMediaProfile class
class SocialMediaProfile:
    def __init__(self, username):
        self.username = username
        self.posts = []

    def add_post(self, post_content):
        self.posts.append(post_content)
        print("Post added successfully!")

    def display_posts(self):
        print(f"Posts by {self.username}:")
        for idx, post in enumerate(self.posts, 1):
            print(f"{idx}. {post}")

    def search_posts(self, keyword):
        matching_posts = [post for post in self.posts if keyword.lower() in post.lower()]

        if matching_posts:
            print(f"Posts containing '{keyword}':")
            for idx, post in enumerate(matching_posts, 1):
                print(f"{idx}. {post}")
        else:
            print(f"No posts found containing '{keyword}'.")

# Function to take input from the user and create a social media profile
def create_social_media_profile():
    username = input("Enter your username: ")
    profile = SocialMediaProfile(username)
    return profile

# Main function to run the program
def main():
    profile = create_social_media_profile()

    while True:
        print("\nMenu:")
        print("1. Add Post")
        print("2. Display Posts")
        print("3. Search Posts")
        print("4. Exit")

        choice = input("Enter your choice (1/2/3/4): ")

        if choice == '1':
            post_content = input("Enter your post: ")
            profile.add_post(post_content)

        elif choice == '2':
            profile.display_posts()

        elif choice == '3':
            keyword = input("Enter a keyword to search for: ")
            profile.search_posts(keyword)

        elif choice == '4':
            print("Exiting program.")
            break

        else:
            print("Invalid choice. Please enter a valid option.")

# Run the main function
if __name__ == "__main__":
    main()
```

# **Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.**

```python
# Importing the datetime module
import datetime

# Creating a class for the ToDo items
class ToDoItem:
    def __init__(self, description, due_date):
        # Setting the attributes of the item
        self.description = description
        self.due_date = datetime.datetime.strptime(due_date, '%Y-%m-%d')
        self.completed = False

    def __str__(self):
        # Defining how the item is displayed
        return f'{self.description} (Due: {self.due_date.strftime("%Y-%m-%d")})'

# Creating a class for the ToDo list
class ToDoList:
    def __init__(self):
        # Initializing an empty list of items
        self.items = []

    def add_item(self, description, due_date):
        # Creating a new item and appending it to the list
        new_item = ToDoItem(description, due_date)
        self.items.append(new_item)
        print(f'Task "{description}" added to the ToDo list.')

    def delete_item(self, index):
        # Deleting an item from the list based on its index
        try:
            # Getting the item to be deleted
            item = self.items[index]
            # Removing the item from the list
            self.items.remove(item)
            print(f'Task "{item.description}" deleted from the ToDo list.')
        except IndexError:
            # Handling the error if the index is invalid
            print('Invalid task index.')

    def complete_item(self, index):
        # Marking an item as completed based on its index
        try:
            # Getting the item to be completed
            item = self.items[index]
            # Setting the completed attribute to True
            item.completed = True
            print(f'Task "{item.description}" marked as completed.')
        except IndexError:
            # Handling the error if the index is invalid
            print('Invalid task index.')

    def show_items(self):
        # Displaying the items in the list
        if self.items:
            print("ToDo List:")
            # Looping through the items and printing them
            for i, item in enumerate(self.items):
                # Adding a check mark if the item is completed
                if item.completed:
                    print(f'[x] {i + 1}. {item}')
                else:
                    print(f'[ ] {i + 1}. {item}')
        else:
            print('No tasks in the ToDo list.')

# Creating an instance of the ToDo list
my_todo_list = ToDoList()

# Creating a menu for the user input
menu = """
1. Add Task
2. Delete Task
3. Complete Task
4. Show Tasks
5. Exit
"""

# Running a loop to get the user input
while True:
    # Printing the menu
    print(menu)
    # Getting the user choice
    choice = input("Enter your choice (1/2/3/4/5): ")

    if choice == '1':
        # Getting the task description and due date from the user
        description = input("Enter task description: ")
        due_date = input("Enter due date (YYYY-MM-DD): ")
        # Calling the add_item method of the ToDo list
        my_todo_list.add_item(description, due_date)

    elif choice == '2':
        # Getting the task index from the user
        index = int(input("Enter the index of the task to delete: "))
        # Calling the delete_item method of the ToDo list
        my_todo_list.delete_item(index - 1)  # Adjusting the index to 0-based

    elif choice == '3':
        # Getting the task index from the user
        index = int(input("Enter the index of the task to complete: "))
        # Calling the complete_item method of the ToDo list
        my_todo_list.complete_item(index - 1)  # Adjusting the index to 0-based

    elif choice == '4':
        # Calling the show_items method of the ToDo list
        my_todo_list.show_items()

    elif choice == '5':
        # Exiting the loop and the program
        print("Exiting the ToDo application. Goodbye!")
        break

    else:
        # Handling the invalid choice
        print("Invalid choice. Please enter a valid option.")
```
